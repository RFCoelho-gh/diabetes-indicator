```python
#import libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import MinMaxScaler
```

```python
# Load the CSV file from the data folder into a DataFrame
main_df = pd.read_csv("data/diabetes_dataset.csv")

# Display the first few rows of the DataFrame
main_df.head(3)
```

```python
main_df.columns
```

```python
main_df.dtypes
```

```python
main_df.isnull().sum()
```

```python
features = main_df.drop(columns=['Diabetes_012'])
```

```python
# Identify columns with a maximum value above 3
columns_to_normalize = [col for col in main_df.columns if main_df[col].max() > 3]

# Apply Min-Max Scaling to the identified columns only
scaler = MinMaxScaler()
main_df[columns_to_normalize] = scaler.fit_transform(main_df[columns_to_normalize])
```

```python
main_df.head(3)
```

```python
target = main_df["Diabetes_012"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.20, random_state=0)
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
knn.fit(X_train, y_train)
```

```python
pred = knn.predict(X_test)
pred
```

```python
knn.score(X_test, y_test)
```
